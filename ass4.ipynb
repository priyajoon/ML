```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def function(x):
    return (x + 3) ** 2

def gradient(x):
    return 2 * (x + 3)  # Derivative of (x + 3)^2 with respect to x

# Gradient Descent parameters
x = 2  # Starting point
learning_rate = 0.1  # Step size
iterations = 1000  # Number of iterations

# Lists to store values of x and y for plotting
x_values = [x]
y_values = [function(x)]

# Perform Gradient Descent
for i in range(iterations):
    grad = gradient(x)  # Calculate the gradient
    x = x - learning_rate * grad  # Update x based on the gradient
    y = function(x)  # Calculate the function value for tracking
    x_values.append(x)
    y_values.append(y)

print(f"\nLocal minimum occurs at x = {x:.4f}, y = {y:.4f}")

# Plotting the function and the Gradient Descent path
x_range = np.linspace(-5, 3, 100)  # X values for the function plot
y_range = function(x_range)  # Y values for the function plot

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label='y = (x + 3)^2', color='blue')  # Function curve
plt.scatter(x_values, y_values, color='red', label='Gradient Descent steps')  # Points visited by Gradient Descent
#plt.plot(x_values, y_values, color='red', linestyle='--')  # Line connecting the steps
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent on y = (x + 3)^2')
plt.grid(True)
plt.legend()
plt.show()
```
